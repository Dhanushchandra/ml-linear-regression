```python
import numpy as np
```

```python
X = np.array([[1],[2],[3],[4],[5],[6]])
y = np.array([1,2,3,4,5,100])
```

```python
Q1 = np.percentile(y,25)
Q3 = np.percentile(y,75)
IQR = Q3 - Q1
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
mask = (y >= lower) & (y <= upper)
```

```python
X_clean = X[mask]
y_clean = y[mask]
```

```python
print("Mask:", mask)
print("Clean X:\n", X_clean)
print("Clean y:", y_clean)
```
