```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([[1], [2], [3], [4], [5], [6]])
y = np.array([2, 3, 4, 5, 6, 100])
```

```python
plt.boxplot(y, vert=False)
plt.title("Boxplot showing IQR and Outliers")
plt.xlabel("y values")
plt.show()
```

```python
Q1 = np.percentile(y, 25)
Q3 = np.percentile(y, 75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

mask = (y >= lower) & (y <= upper)


X_clean = X[mask]
y_clean = y[mask]

print(mask)
```

```python
print(X_clean)
print(y_clean)
```

```python
model = LinearRegression()
model.fit(X_clean,y_clean)
```

```python
print(model.coef_[0])
print(model.intercept_)
```

```python
y_pred = model.predict([[4],[5],[20]])
```

```python
print(X.shape)
print(y_pred)
```

```python
# Plot
# plt.scatter(X, y)
plt.plot([[4],[5],[20]], y_pred)
plt.scatter([[4],[5],[20]], y_pred)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Visualization")
plt.show()
```
